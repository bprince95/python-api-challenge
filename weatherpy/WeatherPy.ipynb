```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import json

from api_keys import API_key
from citipy import citipy

output_data_file = "output_data/cities.csv"

lat_range = (-90, 90)
lng_range = (-180, 180)

url = "http://api.openweathermap.org/data/2.5/weather?"
query_url = f"{url}appid={API_key}&q="

#{city}&units=metric"
```

```python
lat_lngs = []
cities = []

lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    if city not in cities:
        cities.append(city)
len(cities)
```

```python
#print(cities)
```

```python
cityy = []
cloudiness = []
humidity = []
lat = []
lng = []
max_temp = []
wind_speed = []
x=0

for city in cities:
    try:
        x = x+1
        response = requests.get(query_url + city + "&units=imperial")
        response_json = response.json()
        cityy.append(response_json['name'])
        cloudiness.append(response_json['clouds']['all'])
        humidity.append(response_json['main']['humidity'])
        lat.append(response_json['coord']['lat'])
        lng.append(response_json['coord']['lon'])
        max_temp.append(response_json['main']['temp_max'])
        wind_speed.append(response_json['wind']['speed'])
        print(f"Processing record {x} out of {len(cities)} Finding information for: {city}")
    except KeyError:
        print("There is no information for that city. Better luck next time.")
        

    
time.sleep(10)
```

```python
weather_dict = {
    "city": cityy,
    "cloudiness" : cloudiness,
    "humidity" : humidity, 
    "latitude" : lat, 
    "longitude" : lng, 
    "Max_temp" : max_temp, 
    "wind_speed" : wind_speed
}

weather_data = pd.DataFrame(weather_dict)
weather_data

weather_data.to_csv(r'C:\Users\Bowen Prince\Desktop\python-api-challenge\weatherpy\weather_data.csv', index=None)
```

```python
#temp vs. latitude

x = weather_data["Max_temp"]
y = weather_data["latitude"]
plt.xlabel("Maximum Temperature (F)")
plt.ylabel("Latitude")
plt.title("Maximum Temperature (F) vs. Latitude")

plt.scatter(x,y)

plt.savefig('Temp_latitude.png')
```
